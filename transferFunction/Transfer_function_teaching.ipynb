Dynamic activation function plot example:

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact


def linear_activation(x, k):
    return k * x


# Define x range
x = np.linspace(-10, 10, 100)


def plot_activation(k):
    y = linear_activation(x, k)

    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f'Linear Activation: y = {k:.2f}x', color='blue')
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.ylim([-10, 10])  # Fix y-axis limits
    plt.xlabel("Input (x)")
    plt.ylabel("Output (y)")
    plt.title("Dynamic Linear Activation Function")
    plt.legend()
    plt.show()


# Create an interactive slider
interact(plot_activation, k=(-5.0, 5.0, 0.1))
```

```python
def logistic(x, k=1):
    return 1 / (1 + np.exp(-k * x))


def plot_logistic(k):
    y = logistic(x, k)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f'Logistic (k={k})', color='red')
    plt.title("Logistic Activation Function")
    plt.xlabel("Input (x)")
    plt.ylabel("Output (y)")
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.ylim([-2, 2])  # Fix y-axis limits
    plt.legend()
    plt.show()


interact(plot_logistic, k=(-5.0, 5.0, 0.1))
```

```python
def hyperbolicTan(x, k=1):
    return (np.exp(k * x) - np.exp(-k * x)) / (np.exp(k * x) + np.exp(-k * x))


def plot_hyperbolicTan(k):
    y = hyperbolicTan(x, k)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f'Hyperbolic Tangent (k={k})', color='green')
    plt.title("Hyperbolic Tangent Activation Function")
    plt.xlabel("Input (x)")
    plt.ylabel("Output (y)")
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.ylim([-2, 2])  # Fix y-axis limits
    plt.legend()
    plt.show()


interact(plot_hyperbolicTan, k=(-5.0, 5.0, 0.1))
```

```python
def softPlus(x, k=1):
    return np.log(1 + np.exp(k * x))


def plot_softPlus(k):
    y = softPlus(x, k)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f'Softplus (k={k})', color='purple')
    plt.title("Softplus Activation Function")
    plt.xlabel("Input (x)")
    plt.ylabel("Output (y)")
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.ylim([-2, 2])  # Fix y-axis limits
    plt.legend()
    plt.show()


interact(plot_softPlus, k=(-5.0, 5.0, 0.1))
```

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x < 0, alpha * x, x)


def plot_leaky_relu(alpha):
    y = leaky_relu(x, alpha)

    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f'Leaky ReLU (alpha={alpha})', color='purple')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axvline(0, color='black', linestyle='--', linewidth=1)
    plt.xlabel("x")
    plt.ylabel("Leaky ReLU(x)")
    plt.title("Leaky ReLU Funkce s parametrem Î±")
    plt.legend()
    plt.show()


interact(plot_leaky_relu, alpha=(0.0, 1.0, 0.05))
```

```python
def gaussian(x, k=1):
    return np.exp(-k * x  2)


def plot_gaussian(k):
    y = gaussian(x, k)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f'Gaussian (k={k})', color='orange')
    plt.title("Gaussian Activation Function")
    plt.xlabel("Input (x)")
    plt.ylabel("Output (y)")
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.ylim([-2, 2])  # Fix y-axis limits
    plt.legend()
    plt.show()


interact(plot_gaussian, k=(0.1, 5.0, 0.1))
```

```python
def softmax(x, k=1):
    return np.exp(k * x) / np.sum(np.exp(k * x))


def plot_softmax(k):
    y = softmax(x, k)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=f'Softmax (k={k})', color='orange')
    plt.title("Softmax Activation Function")
    plt.xlabel("Input (x)")
    plt.ylabel("Output (y)")
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.axvline(0, color='black', linewidth=1, linestyle='--')
    plt.ylim([-2, 2])  # Fix y-axis limits
    plt.legend()
    plt.show()


interact(plot_softmax, k=(0.1, 5.0, 0.1))
```

# Activation Functions

## Introduction
Activation functions play a crucial role in neural networks by introducing non-linearity, which enables them to learn complex patterns. This project explores five different activation functions: Logistic (Sigmoid), Hyperbolic Tangent, Softplus, Leaky ReLU, and Gaussian, with a parameter \( k \) (or \( \alpha \) in the case of Leaky ReLU) that modifies their behavior.

Below, we analyze their use cases, advantages, and disadvantages.


## 1. Logistic (Sigmoid) Activation Function
### Equation:
\[
f(x) = \frac{1}{1 + e^{-k x}}
\]

### Use Cases:
- Common in binary classification tasks (e.g., logistic regression).
- Historically used in neural networks before ReLU became popular.

### Advantages:
âœ” Smooth and differentiable.  
âœ” Output is always in the range (0,1), making it useful for probability estimation.  

### Disadvantages:
âœ– Prone to vanishing gradient problem, slowing down training.  
âœ– Output saturates for large positive/negative values of \( x \).  


## 2. Hyperbolic Tangent (Tanh) Activation Function
### Equation:
\[
f(x) = \frac{e^{k x} - e^{-k x}}{e^{k x} + e^{-k x}}
\]

### Use Cases:
- Used in hidden layers of neural networks, especially before ReLU gained popularity.
- Suitable when outputs need to be centered around zero.

### Advantages:
âœ” Centered at 0, which helps in faster training compared to the sigmoid.  
âœ” Stronger gradients than the logistic function.  

### Disadvantages:
âœ– Also suffers from the vanishing gradient problem for large values of \( |x| \).  
âœ– Computationally more expensive than ReLU.  


## 3. Softplus Activation Function
### Equation:
\[
f(x) = \ln(1 + e^{k x})
\]

### Use Cases:
- Smooth alternative to ReLU, used in certain specialized deep learning models.
- Can be seen as a "soft" version of ReLU, ensuring differentiability.

### Advantages:
âœ” Always differentiable and smooth.  
âœ” Avoids the dying ReLU problem (where neurons output zero permanently).  

### Disadvantages:
âœ– More computationally expensive than ReLU.  
âœ– Similar to ReLU but without significant performance advantages.  


## 4. Leaky ReLU Activation Function
### Equation:
\[
f(x) =
\begin{cases} 
\alpha x, & x < 0 \\
x, & x \geq 0
\end{cases}
\]

### Use Cases:
- Overcomes the dying ReLU problem by allowing small gradients for negative values.
- Common in deep learning architectures.

### Advantages:
âœ” Prevents neurons from being permanently inactive (dying ReLU).  
âœ” Computationally efficient and simple to implement.  

### Disadvantages:
âœ– Choosing \(\alpha\) is tricky; too high values distort learning.  
âœ– Still suffers from exploding gradients in deep networks.  


## 5. Gaussian Activation Function
### Equation:
\[
f(x) = e^{-k x^2}
\]

### Use Cases:
- Less common in deep learning but useful in radial basis function (RBF) networks.
- Used in feature transformation and certain kernel-based methods.

### Advantages:
âœ” Provides localized activation, useful in probabilistic models.  
âœ” Can model non-linear relationships effectively.  

### Disadvantages:
âœ– Not widely used in standard deep learning frameworks.  
âœ– Can cause vanishing gradients for large values of \( k \).  


## Summary: Reflections and Confusions
### Key Takeaways:
- Adding the parameter \( k \) allowed us to control the shape and behavior of each activation function.
- Some functions, like Softplus and Leaky ReLU, serve as alternatives to ReLU to avoid its drawbacks.
- The Gaussian function behaves very differently from others, making it less commonly used in neural networks.

### Biggest Confusion: Incorporating `k`
The most challenging part was how to integrate `k` into each function correctly:
- For Logistic, Tanh, and Softplus, `k` was a scaling factor for \( x \).
- In Leaky ReLU, `k` was replaced by \(\alpha\) to modify negative values.
- For Gaussian, `k` controlled the spread of the function.
